```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=np.array([1,2,3,4])
y=np.array([10,60,40,100])
```

```python
plt.plot(x,y,color="r",marker="o",ms="10",mec="0",mfc="b")
plt.grid(color="g")
```

```python
plt.plot(x,y,'og')
```

```python
plt.plot(x,y,ls='dashed',linewidth="2",color="r",marker="o",ms="15",mec="0",mfc="b")
plt.grid(color="g")
```

```python
x1=np.array([10,30,60,120,150])
y1=np.array([30,50,70,120,250])
x2=np.array([20,60,80,130,190])
y2=np.array([40,80,160,220,250])
plt.plot(x1,y1,x2,y2,ls='-.',linewidth='3',marker="D",ms="6",mec="0",mfc="m")
plt.grid()
plt.show()
```
